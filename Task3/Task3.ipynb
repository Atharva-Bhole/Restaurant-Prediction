```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("../Dataset.csv")
df.head()
```

```python
df.isna().sum()
```

```python
df.dropna(inplace=True)
df.shape
```

```python
cuisines_dummies = df['Cuisines'].str.get_dummies(sep=', ')
cuisines_dummies
```

```python
data = pd.concat([df, cuisines_dummies], axis='columns')
data
```

```python
data.drop(columns=['Restaurant ID', 'Latitude', 'Longitude', 'City', 'Address', 'Locality', 'Locality Verbose', 'Cuisines', 'Currency', 'Has Table booking', 'Has Online delivery', 'Is delivering now', 'Switch to order menu', 'Rating color', 'Rating text'], inplace=True)
```

```python
data
```

```python
X = data.drop(columns='Restaurant Name')
y = data['Restaurant Name']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=30)
```

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
model.fit(X, y)
model.score(X_test, y_test)
```
