```python
import pandas as pd
df = pd.read_csv('../Dataset.csv')
df.head()
```

```python
df.isna().sum()
```

```python
df.dropna(inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder, StandardScaler
le = LabelEncoder()
scaler = StandardScaler()
```

```python
curr = le.fit_transform(df['Currency'])
curr
```

```python
names = le.fit_transform(df['Restaurant Name'])
vote = scaler.fit_transform(df[['Votes']])
```

```python
cuisines = le.fit_transform(df['Cuisines'])
```

```python
cuisines_dummies = df['Cuisines'].str.get_dummies(sep=', ')
cuisines_dummies
```

```python
cuisines_dummies.replace({0:False, 1:True}, inplace=True)
```

```python
df_with_dummies = pd.concat([df, cuisines_dummies], axis='columns')
df_without_dummies = df
df_without_dummies['Cuisines'] = cuisines
```

```python
df_without_dummies
```

```python
df_with_dummies
```

```python
df_with_dummies['Restaurant Name'] = names
df_without_dummies['Restaurant Name'] = names
```

```python
df_with_dummies.drop(columns=['Restaurant ID', 'City', 'Address', 'Locality', 'Locality Verbose', 'Latitude', 'Longitude', 'Rating color', 'Rating text', 'Cuisines', 'World Cuisine'], inplace=True)
df_without_dummies.drop(columns=['Restaurant ID', 'City', 'Address', 'Locality', 'Locality Verbose', 'Latitude', 'Longitude', 'Rating color', 'Rating text'], inplace=True)
```

```python
df_with_dummies['Has Online delivery'].replace({'Yes':True, 'No':False}, inplace=True)
df_with_dummies['Has Table booking'].replace({'Yes':True, 'No':False}, inplace=True)
df_with_dummies['Is delivering now'].replace({'Yes':True, 'No':False}, inplace=True)
df_with_dummies['Switch to order menu'].replace({'Yes':True, 'No':False}, inplace=True)
df_without_dummies['Has Online delivery'].replace({'Yes':True, 'No':False},inplace=True)
df_without_dummies['Has Table booking'].replace({'Yes':True, 'No':False}, inplace=True)
df_without_dummies['Is delivering now'].replace({'Yes':True, 'No':False}, inplace=True)
df_without_dummies['Switch to order menu'].replace({'Yes':True, 'No':False}, inplace=True)
```
