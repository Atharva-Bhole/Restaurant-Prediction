```python
import pandas as pd
```

```python
df = pd.read_csv("../Dataset.csv")
df
```

```python
na_rest = df[df['Cuisines'].isna()]
na_rest
```

We dropped the null value rows because we cannot decide what may be their cuisines

```python
df.dropna(inplace=True)
df['Restaurant Name'].unique().shape
```

Filtering the data and Balancing the data

```python
df['Restaurant Name'].unique().shape
```

```python
train_data = df.drop_duplicates("Restaurant Name")
train_data
```

Dropping the unwanted features and selecting most important features

```python
train_data.drop(columns=['Restaurant ID', 'City', 'Address', 'Locality', 'Locality Verbose', 'Latitude', 'Longitude', 'Currency', 'Rating color', 'Rating text'], inplace=True)
```

```python
train_data
```

Label Encoding the data

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
names = le.fit_transform(train_data['Restaurant Name'])
names
```

```python
names_dict = dict(zip(names, train_data['Restaurant Name']))
names_dict
```

```python
len(names_dict)
```

```python
cuisines = train_data['Cuisines'].str.get_dummies(sep=', ')
cuisines
```

```python
final_proc_data = pd.concat([train_data, cuisines], axis='columns')
final_proc_data
```

```python
final_proc_data['Restaurant Name'] = names
final_proc_data
```

```python
final_proc_data['Has Online delivery'].replace({'Yes':True, 'No':False}, inplace=True)
final_proc_data['Has Table booking'].replace({'Yes':True, 'No':False}, inplace=True)
final_proc_data['Is delivering now'].replace({'Yes':True, 'No':False}, inplace=True)
final_proc_data['Switch to order menu'].replace({'Yes':True, 'No':False}, inplace=True)
final_proc_data
```

```python
final_proc_data.drop(columns='Cuisines', inplace=True)
final_proc_data
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
final_proc_data['Votes'] = scaler.fit_transform(final_proc_data[['Votes']])
```

```python
final_proc_data
```

```python
final_proc_data['Votes'].std()
```

```python
X = final_proc_data.drop(columns='Restaurant Name')
y = final_proc_data['Restaurant Name']
X
```

```python
y
```

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
model.fit(X,y)
```

```python
model.score(X,y)
```
