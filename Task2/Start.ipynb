```python
import pandas as pd
df = pd.read_csv('../Dataset.csv')
```

```python
df = df.dropna()
```

```python
df["Aggregate rating"].unique()
```

```python
df['Aggregate rating'] = round(df['Aggregate rating'])
```

```python
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
le = LabelEncoder()
minmmaxscaler = MinMaxScaler()
curr = le.fit_transform(df['Currency'])
ratext = le.fit_transform(df['Rating text'])
ratecolor = le.fit_transform(df['Rating color'])
vote = minmmaxscaler.fit_transform(df[['Votes']])
```

```python
df['Aggregate rating'].unique()
```

```python
df['Votes'] = vote
df['Rating color'] = ratecolor
df['Rating text'] = ratext
df['Currency'] = curr
```

```python
df
```

```python
cuisine_dumm = df['Cuisines'].str.get_dummies(sep=', ')
cuisine_dumm
```

```python
cuisine_dumm = cuisine_dumm.replace({0: False, 1 : True})
```

```python
cuisine_dumm
```

```python
df_final = pd.concat([df, cuisine_dumm], axis='columns')
df_final
```
